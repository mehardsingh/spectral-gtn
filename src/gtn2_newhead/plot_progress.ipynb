```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../../checkpoints/gtn2/discrete/progress.csv")
df
```

```python
epochs = df["epoch"]
t_loss = df["t_loss"]
t_acc = df["t_acc"]
t_f1 = df["t_f1"]
v_loss = df["v_loss"]
v_acc = df["v_acc"]
v_f1 = df["v_f1"]
```

```python
plt.plot(epochs, t_loss, label='Training Loss')
plt.plot(epochs, v_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(epochs, t_acc, label='Training Acc')
plt.plot(epochs, v_acc, label='Validation Acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(epochs, t_f1, label='Training F1')
plt.plot(epochs, v_f1, label='Validation F1')
plt.xlabel('Epoch')
plt.ylabel('F1')
plt.legend()
plt.show()
```
